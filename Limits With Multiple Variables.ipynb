Here are 2 functions f and g. For this script I will be finding an $m$ and $n$ such that the limit as x -> c exist. This means that the limit as

$ f(x) = \begin{cases} 
      mx & x > c \\
      nx^{2} & x < c\\ 
   \end{cases}  $


So let's a try problem. <br>


Let $c=3$ <br>
We want $\lim_{x \to 3} f(x)$ to exist.
See what happens if m = 1 and n=1.





```python
import matplotlib.pyplot as plt
import numpy as np


def plotPieceLimit(_m,_n,_c,):
    m = _m      # m that we are testing
    n = _n      # n that we are testing
    c = _c      # c is where x is approching in limit


    fRight = "n*xRight"
    fLeft = "m*xLeft**2"
    xLeft = np.linspace(c-5,c, 100)
    xRight = np.linspace(c,c+5, 100)





    yLeft = eval(fLeft)
    yRight = eval(fRight)


    plt.plot(xLeft, yLeft)
    plt.plot(xRight, yRight)
    plt.plot([c,c],[np.amin(np.concatenate((yLeft,yRight))), np.amax(np.concatenate((yLeft,yRight))) ], linestyle='dashed')
    
plotPieceLimit(1,1,3)
```

So obviously when m = 1 the limit DNE at c = 3 <br>
Mathematically this is expressed as, <br>
$\lim_{x \to 3} f(x)  = \lim_{x \to 3} \begin{cases} 
      mx & x > c \\
      mx^{2} & x < c\\ 
   \end{cases} = \lim_{x \to 3} \begin{cases} 
      1*x & x > 3 \\
      1*x^{2} & x < 3\\ 
   \end{cases} = DNE$

Doing some calculation you can find the m value such that $\lim_{x \to 3^{-}} f(x) = \lim_{x \to 3^{+}} f(x)$

$m*c^{2} = n*c$ <br>
$m*c = n$

What this tells us is for any chosen $m$ we can find an $n$ to ensure the limit of $f$ to exist at $c$

So if $m=1$ <br>
Then $n = c*m = 3*1 = 3$ <br>
Let's try it...

```python
plotPieceLimit(1,3,3)
```

```python
plotPieceLimit(2,6,3)
```

```python
plotPieceLimit(3,3*7,7)
```
